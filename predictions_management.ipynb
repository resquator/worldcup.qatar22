```python
import pandas as pd
import numpy as np 
import os

```

```python
from os import listdir
from os.path import isfile, join
mypath = 'games_predict/'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
files = [f for f in onlyfiles if 'g_' in f]

full = pd.DataFrame()
for f in files:
    data = pd.read_csv(f'games_predict/{f}', delimiter=',')
    #home_team,away_team,home_score,away_score,h_points,a_points,timestamp
    data.columns = ['idx','home_team', 'away_team', 'home_score','away_score','h_points','a_points','timestamp']
    full = pd.concat([full, data], axis=0)
    

```

```python
full=full.reset_index()
full
```

```python
for idxrow, r in full.iterrows():
    h=str(r.home_score)
    a=str(r.away_score)
    full.loc[idxrow,'correct_score']=h+':'+a

full.query('home_team == "England" and away_team == "Iran"')    
```

```python
full
```

```python
one_game=full
one_game.groupby(['home_team','away_team','correct_score']).agg({'idx':'count'}).sort_values('idx', ascending=False)
```
